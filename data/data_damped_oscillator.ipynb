Create a damped oscillator with a given damping factor and frequency.

```python
# import libraries
import numpy as np 
import pandas as pd
from scipy.integrate import solve_ivp
import matplotlib.pyplot as plt
```

```python
t = np.linspace(0,30,300)
y = [0, 1]
# damping constant
gamma = 0.05
# frequency of oscillation
omega_sqr = 5

# function to solve
def sho(t,y):
    solution = (y[1],(-gamma*y[1]-omega_sqr*y[0]))
    return solution

solution = solve_ivp(sho, [0,300], y0 = y, t_eval = t)
```

End of code citation.

Get the x and y data from the damped oscillator data:

```python
x_damped_oscillator = solution.t
y_damped_oscillator = solution.y[0]
```

Plot the data:

```python
plt.plot(x_damped_oscillator, y_damped_oscillator)

plt.ylabel("y")
plt.xlabel("x")
plt.title('Data: Damped Oscillator', fontsize = 15)

plt.show()
```

```python
dataset = pd.DataFrame({"x":x_damped_oscillator, "y":y_damped_oscillator})

# safe dataset
dataset.to_csv("data_damped_oscillator.csv", index = False)
```
